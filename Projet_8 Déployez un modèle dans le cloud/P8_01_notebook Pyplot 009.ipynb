```python
import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

from sklearn.cluster import KMeans
```

```python
data_pca_2D = pd.read_csv('03_data_csv/data_sample_009/df_features_pca_affichage2D.csv', delimiter=";", index_col=0)
data_categ = pd.read_csv('03_data_csv/data_sample_009/df_categ.csv', delimiter=";", index_col=0)
data_tsne = pd.read_csv('03_data_csv/data_sample_009/df_features_tsne_affichage2D.csv', delimiter=";", index_col=0)
data_feat = pd.read_csv('03_data_csv/data_sample_009/df_features_pca.csv', delimiter=";", index_col=0)
```

```python
km_acp = KMeans(n_clusters=131)
km_acp.fit(data_feat)
```

# PCA

```python
df =data_pca_2D.join(data_categ)
```

```python
px.scatter(df, x="X", y="Y",color= km_acp.labels_, hover_name = "categ")
```

```python
px.scatter(df, x="X", y="Y",color="categ", hover_name =  km_acp.labels_)
```

# TSNE

```python
df =data_tsne.join(data_categ)
```

```python
px.scatter(df, x="tsne1", y="tsne2",color= km_acp.labels_, hover_name =  "categ")
```

```python
px.scatter(df, x="tsne1", y="tsne2",color="categ", hover_name = km_acp.labels_)
```
