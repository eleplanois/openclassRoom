# constitution d'un seul datase à partir de la moyenne de 2015 et 2016

```python
data_moyenne = df2015.merge(df2016, on='OSEBuildingID', how='inner', suffixes=['_2015','_2016'])
data_moyenne.head()
```

```python
colonne= ['ENERGYSTARScore','SiteEnergyUseWN(kBtu)', 'NumberofFloors', \
          'LargestPropertyUseTypeGFA', 'PropertyGFAParking', 'SiteEUIWN(kBtu/sf)',\
         'SteamUse(kBtu)','Electricity(kBtu)','NaturalGas(kBtu)']

for col in colonne:
    data_moyenne[col+'_mean'] = ((data_moyenne[col+'_2015']+data_moyenne[col+'_2016'])/2)
```

```python
for col in colonne:
    filter_2016=data_moyenne[col+'_2015'].isna() & data_moyenne[col+'_2016'].notna()
    filter_2015=data_moyenne[col+'_2015'].notna() & data_moyenne[col+'_2016'].isna()
    data_moyenne.loc[filter_2016,col+'_mean'] = data_moyenne.loc[filter_2016,col+'_2016']
    data_moyenne.loc[filter_2015,col+'_mean'] = data_moyenne.loc[filter_2015,col+'_2015']
```

```python
for col in colonne:
    print('valeurs manquantes pour colonne ',col)
    print('   2015 :',data_moyenne[col+'_2015'].isna().sum()/len(data_moyenne)*100)
    print('   mean :',data_moyenne[col+'_mean'].isna().sum()/len(data_moyenne)*100, '\n')
```

```python
data_moyenne['GHGEmissions(MetricTonsCO2e)_mean'] = ((data_moyenne['GHGEmissions(MetricTonsCO2e)']+data_moyenne['TotalGHGEmissions'])/2)
```

```python
data_moyenne[data_moyenne['GHGEmissions(MetricTonsCO2e)_mean'].isna()][['GHGEmissions(MetricTonsCO2e)_mean','GHGEmissions(MetricTonsCO2e)','TotalGHGEmissions']].head(100)
```

```python
filter_2016=data_moyenne['GHGEmissions(MetricTonsCO2e)'].isna() & data_moyenne['TotalGHGEmissions'].notna()
filter_2015=data_moyenne['GHGEmissions(MetricTonsCO2e)'].notna() & data_moyenne['TotalGHGEmissions'].isna()
data_moyenne.loc[filter_2016,'GHGEmissions(MetricTonsCO2e)_mean'] = data_moyenne.loc[filter_2016,'TotalGHGEmissions']
data_moyenne.loc[filter_2015,'GHGEmissions(MetricTonsCO2e)_mean'] = data_moyenne.loc[filter_2015,'GHGEmissions(MetricTonsCO2e)']
```

```python
data_moyenne[data_moyenne['GHGEmissions(MetricTonsCO2e)_mean'].isna()][['GHGEmissions(MetricTonsCO2e)_mean','GHGEmissions(MetricTonsCO2e)','TotalGHGEmissions']].head(100)
```

```python
data_moyenne['GHGEmissionsIntensity(kgCO2e/ft2)_mean'] = ((data_moyenne['GHGEmissionsIntensity(kgCO2e/ft2)']+data_moyenne['GHGEmissionsIntensity'])/2)
data_moyenne[data_moyenne['GHGEmissionsIntensity(kgCO2e/ft2)_mean'].isna()][['GHGEmissionsIntensity(kgCO2e/ft2)_mean','GHGEmissionsIntensity(kgCO2e/ft2)','GHGEmissionsIntensity']].head(100)
```

```python
for col in colonne:
    colu=col+'_mean'
    print(f'{col} colonne vide {data_moyenne[colu].isna().sum()/len(data_moyenne)*100}') 
```

### calcul composition consommation energie

```python
dic_typ_energy = {}
for typ in data_moyenne['PrimaryPropertyType_2015'].unique():
    steam = data_moyenne.loc[(data_moyenne['PrimaryPropertyType_2015']==typ), 'SteamUse(kBtu)_mean'].mean()
    elec = data_moyenne.loc[(data_moyenne['PrimaryPropertyType_2015']==typ), 'Electricity(kBtu)_mean'].mean()
    gaz = data_moyenne.loc[(data_moyenne['PrimaryPropertyType_2015']==typ), 'NaturalGas(kBtu)_mean'].mean()
    total = steam + elec + gaz
    dic_typ_energy[typ] = [ steam/total, gaz/total, elec/total]
    
```

```python
data_moyenne['PrimaryPropertyType_2015'].apply(lambda x: dic_typ_energy[x][0])
data_moyenne['util_steam'] = data_moyenne['PrimaryPropertyType_2015'].apply(lambda x: dic_typ_energy[x][0])
data_moyenne['util_elec'] = data_moyenne['PrimaryPropertyType_2015'].apply(lambda x: dic_typ_energy[x][1])
data_moyenne['util_gaz'] = data_moyenne['PrimaryPropertyType_2015'].apply(lambda x: dic_typ_energy[x][2])
```

```python
data_moyenne['age']=2015 - data_moyenne.YearBuilt_2015
```

```python
data_moyenne.head()
```

```python
#votre équipe s’intéresse de près aux émissions des bâtiments non destinés à l’habitation.
data_moyenne.drop(index = data_moyenne[data_moyenne['PrimaryPropertyType_2015']=='Low-Rise Multifamily'].index, inplace=True)
data_moyenne.drop(index = data_moyenne[data_moyenne['PrimaryPropertyType_2015']=='Mid-Rise Multifamily'].index, inplace=True)
data_moyenne.drop(index = data_moyenne[data_moyenne['PrimaryPropertyType_2015']=='High-Rise Multifamily'].index, inplace=True)
```
